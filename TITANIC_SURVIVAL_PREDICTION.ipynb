```python
#Mapping text data to binary Value
z2 = set(concat_df3['Sex'])
concat_df3['Sex'] = concat_df3['Sex'].map({'female' : 0, 'male' : 1}).astype(int)
print(concat_df3.head)
```

```python
# Assuming concat_df3 is your DataFrame and you want to remove 'Cabin' and 'Embarked' columns
columns_to_remove = ['Cabin', 'Embarked']

# Check if the columns exist before dropping
columns_exist = all(col in concat_df3.columns for col in columns_to_remove)

if columns_exist:
    concat_df3 = concat_df3.drop(columns=columns_to_remove, axis=1)
else:
    print("Columns to remove not found in DataFrame.")

# Print the modified DataFrame
print(concat_df3)
```

```python

x = concat_df3.drop(['Survived', 'Name', 'Ticket'], axis=1)

x
```

```python
y = concat_df3.Survived
y
```

```python
y = concat_df3['Survived'].fillna(1.0)
y
```

```python
# Finding and removing Null values
x.columns[x.isna().any()]
```

```python
x.Age = x.Age.fillna(x.Age.mean())
x.Fare = x.Fare.fillna(x.Fare.mean())
```

```python
# Finding and removing Null values
x.columns[x.isna().any()]
```

Splitting dataset into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Training
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```
