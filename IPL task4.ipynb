```python
# for manipulation
import numpy as np
import pandas as pd
# for visualization
import matplotlib.pyplot as plt
import seaborn as sns
# for interactiveness
from ipywidgets import interact
```

```python
# lets read data
d = pd.read_csv('Desktop/matches.csv')
d.head(5)
```

```python
# lets us check null values
d.isnull().sum()
```

```python
# Drop rows with missing values
d.dropna(inplace=True)

```

```python
# Check for missing values again
d.isnull().sum()
```

```python
# Check for duplicates
d.duplicated().sum()
```

```python
# lets check the shape of the dataset
d.shape
```

```python
# Calculate summary statistics
d.describe()
```

```python
# Visualize the data using histograms
d.hist(bins=8, figsize=(20,15))
```

```python
 #Visualize the data using scatterplots

fig, ax = plt.subplots(figsize=(10,60))
sns.scatterplot(x='win_by_wickets', y='player_of_match', data = d)

```

```python
# Visualize the data using boxplots
fig, ax = plt.subplots(figsize=(20, 40))
sns.boxplot(x='winner', y='win_by_runs', data=d)
```

```python
# Analyze the data by player
d.groupby('player_of_match')[['win_by_runs', 'win_by_wickets']].sum().plot(kind='bar', figsize=(10,10))
```

```python
# Analyze the data by team
d.groupby('winner')[['win_by_runs', 'win_by_wickets']].sum().plot(kind='bar', figsize=(10,5))
```
